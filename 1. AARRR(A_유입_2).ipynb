```python
# 주문 날짜를 datetime 형식으로 변환
order['order_purchase_timestamp'] = pd.to_datetime(order['order_purchase_timestamp'])

# 연도별로 주문 데이터 분리
order_2017 = order[order['order_purchase_timestamp'].dt.year == 2017]
order_2018 = order[order['order_purchase_timestamp'].dt.year == 2018]

# 각 도시(주)의 주문 건수를 비율로 계산
order_distribution_by_region_2017 = order_2017['customer_state'].value_counts(normalize=True).to_dict()
order_distribution_by_region_2018 = order_2018['customer_state'].value_counts(normalize=True).to_dict()

# 결과 출력
print("2017년 도시(주)별 주문 건수 비율:")
print(order_distribution_by_region_2017)

print("\n2018년 도시(주)별 주문 건수 비율:")
print(order_distribution_by_region_2018)
```

```python
# 도시(주)별 주문 건수 비율과 인구 비율을 비교
for city in city_order_counts.index:
    order_ratio = order_distribution_by_state.get(city, 0)
    population_ratio = city_population_ratio.get(city, 0)
    print(f"{city}: 주문 건수 비율 - {order_ratio:.2f}, 인구 비율 - {population_ratio:.2f}")
```

```python
# 도시(주)별 주문 건수 비율과 인구 비율을 비교
significant_difference_cities = []

for city in city_order_counts.index:
    order_ratio = order_distribution_by_state.get(city, 0)
    population_ratio = city_population_ratio.get(city, 0)

    if population_ratio > 0 and (order_ratio / population_ratio >= 1.5 or population_ratio / order_ratio >= 1.5):
        significant_difference_cities.append((city, order_ratio, population_ratio))

for city, order_ratio, population_ratio in significant_difference_cities:
    print(f"{city}: 주문 건수 비율 - {order_ratio:.2f}, 인구 비율 - {population_ratio:.2f}")
```

```python
# significant_difference_cities 리스트 생성
significant_difference_cities = []

for city in city_order_counts.index:
    order_ratio = order_distribution_by_state.get(city, 0)
    population_ratio = city_population_ratio.get(city, 0)

    if population_ratio > 0 and (order_ratio / population_ratio >= 1.5 or population_ratio / order_ratio >= 1.5):
        significant_difference_cities.append((city, order_ratio, population_ratio))

# 도시, 주문 건수 비율, 인구 비율 리스트 생성
cities = [city for city, order_ratio, population_ratio in significant_difference_cities]
order_ratios = [order_ratio for city, order_ratio, population_ratio in significant_difference_cities]
population_ratios = [population_ratio for city, order_ratio, population_ratio in significant_difference_cities]

# 막대 차트 시각화
x = np.arange(len(cities))  # 도시 인덱스
width = 0.35  # 막대 너비

fig, ax = plt.subplots(figsize=(14, 8))

# 주문 건수 비율 막대
rects1 = ax.bar(x - width/2, order_ratios, width, label='주문 건수 비율')
# 인구 비율 막대
rects2 = ax.bar(x + width/2, population_ratios, width, label='인구 비율')

# 제목 및 라벨 설정
ax.set_xlabel('도시')
ax.set_ylabel('비율')
ax.set_title('주별 주문 건수 비율과 인구 비율 비교')
ax.set_xticks(x)
ax.set_xticklabels(cities, rotation=45, ha='right')
ax.legend()

# 막대에 비율 값 표시
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 텍스트를 약간 위로 이동
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
